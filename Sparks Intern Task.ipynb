# Prediction using Supervised Learning

### Statement: Predicting the score of a student based on the number of hours studied.

```python
#importing the required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the csv file from the given link
data = pd.read_csv("http://bit.ly/w-data")
```

```python
#display the data set
data.head(10)
```

```python
#plot the data set on a scatter plot to visulaize the data set
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.scatter(data["Hours"],data["Scores"])
plt.show()
```

```python
#Divide data as features and label data
X = data.iloc[:,:-1].values
y = data.iloc[:,1].values
```

```python
#split the data as train and test data
from sklearn.model_selection import train_test_split 
X_train , X_test, y_train, y_test = train_test_split(X,y)
```

```python
#create a linear model
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
```

```python
#fit the data to the model
lin.fit(X_train,y_train)
```

```python
#plot the linear regression line on the data set
line = lin.coef_*X+lin.intercept_
plt.scatter(X,y)
plt.plot(X,line,color="red")
plt.show()
```

```python
#preidct the scores on test data
pred = lin.predict(X_test)
df = pd.DataFrame({"Actual Score":y_test,"Predicted Score":pred})
df.head(10)
```

```python
#Calculate the accuracy score
from sklearn.metrics import r2_score,mean_absolute_error
print("Accuracy:",r2_score(y_test,pred))
```

```python
#calculate mean absolute error
print("Mean Absolute Error:",mean_absolute_error(y_test,pred))
```

```python
#predicting the score of the student if he studies for 9.25 hours
hours = 9.25
score = lin.predict([[hours]])
print("The predicted score if the student study for" , hours,"is",score[0])
```
